# feature engineering

* we will do some feature extraction and engineering to our dataset:
    - prepare data to regression by adding ttf label to train set
    - prepare data to classification by adding classifier label to train and test set for binary classification and multiclass classification

- regressiong answer the question: what is the expected time to failure (ttf) ?
- binary classification answer the question: is the machine going to fail in the a specific period of cycles ?
- multiclass classification answer the question: which period will be the most likely to fail ?

```python
def add_feature(df_in, rolling_window):
    """
    args: 
        df_in: input dataframe
        rolling_window: rolling window period 
        
    returns: 
        df_out: output dataframe with added features
    """
    
    sensor_feature = ['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8', 's9', 's10', 's11', 's12', 's13', 's14', 's15', 's16', 's17', 's18', 's19', 's20', 's21']
    
    sensor_avg = [col.replace('s', 'avg') for col in sensor_feature]
    sensor_std = [col.replace('s', 'std') for col in sensor_feature]
    
    df_out = pd.DataFrame()
    rw = rolling_window
    
    for ids in df_in.id.unique():
        df_engine = df_in[df_in.id == ids]
        df_sub = df_engine[sensor_feature]
        
        # rolling mean
        av = df_sub.rolling(window=rw, min_periods=1).mean()
        av.columns = sensor_avg
        
        # rolling std
        std = df_sub.rolling(window=rw, min_periods=1).std().fillna(0)
        std.columns = sensor_std

        # concatinate dataframe
        df_new = pd.concat([df_engine, av, std], axis=1)
        df_out = pd.concat([df_out, df_new])
        
    return df_out
```

```python
def prepare_train_df(df_in, period):
        
    """
    Add regression and classification labels to the training data.

        Regression label: ttf (time-to-failure) = each cycle for an engine subtracted from the last cycle of the same engine
        Binary classification label: label_bnc = if ttf is <= parameter period then 1 else 0. (values = 0,1)
        Multi-class classification label: label_mcc = 2 if ttf <= 0.5 * parameter period , 1 if ttf <= parameter period, else 0. (values = 0,1,2)
        
    Args:
        df_in (dataframe): The input training data
        period (int)     : The number of cycles for TTF segmentation. Used to derive classification labels
        
    Returns:
        dataframe: The input dataframe with regression and classification labels added

    
    illustrate the use of this function reset_index(level=0):     
    
                    Data
    Year Month      
    2000 Jan         1
        Feb         2
    2001 Jan         3
        Feb         4

    df.reset_index(level=0)

        Year  Data
    Month            
    Jan    2000     1
    Feb    2000     2
    Jan    2001     3
    Feb    2001     4
    """

    # regression label
    df_last_cycle = pd.DataFrame(df_in.groupby('id')['cycle'].max()).reset_index(level=0)
    df_last_cycle.columns = ['id', 'last_cycle']
    df_in = pd.merge(df_in, df_last_cycle, on='id')
    
    df_in['ttf'] = df_in['last_cycle'] - df_in['cycle']
    df_in = df_in.drop(columns=['last_cycle'], axis=1)
    
    # binary classification label
    df_in['bnc_label'] = df_in['ttf'].apply(lambda x: 1 if x <= period else 0) # values = 0,1. 0 = no failure, 1 = failure
    
    # multi-class classification label
    df_in['mcc_lable'] = df_in['ttf'].apply(lambda x: 2 if x <= 0.5 * period else 1 if x <= period else 0) # values = 0,1,2. 0 = no failure, 1 = low failure, 2 = high failure
    
    return df_in        
```

```python
def prepare_test_df(df_test, df_truth, period):
    """
    Add regression and classification labels to the training data.

        Regression label: ttf (time-to-failure) = each cycle for an engine subtracted from the last cycle of the same engine
        Binary classification label: label_bnc = if ttf is <= parameter period then 1 else 0. (values = 0,1)
        Multi-class classification label: label_mcc = 2 if ttf <= 0.5 * parameter period , 1 if ttf <= parameter period, else 0. (values = 0,1,2)
        
    Args:
        df_test (dataframe): The input test data
        df_truth (dataframe): The input truth data  
        period (int)     : The number of cycles for TTF segmentation. Used to derive classification labels
        
    Returns:
        dataframe: The input dataframe with regression and classification labels added
    
    """    
    
    # regression label
    df_lst_test_cycle = pd.DataFrame(df_test.groupby('id')['cycle'].max())
    df_lst_test_cycle = df_lst_test_cycle.reset_index(level=0)
    df_lst_test_cycle.columns = ['id', 'last_cycle']
    df_test = pd.merge(df_test, df_lst_test_cycle, on='id')
    
    df_test = df_test[df_test['cycle'] == df_test['last_cycle']]
    df_test = df_test.reset_index(drop=True)
    df_test = df_test.drop(columns=['last_cycle'], axis=1)
    
    df = pd.concat([df_test, df_truth], axis=1)
    
    # binary classification label
    df['bnc_label'] = df['ttf'].apply(lambda x: 1 if x <= period else 0) # values = 0,1. 0 = no failure, 1 = failure
    
    # multi-class classification label
    df['mcc_label'] = df['ttf'].apply(lambda x: 2 if x <= 0.5 * period else 1 if x <= period else 0) # values = 0,1,2. 0 = no failure, 1 = low failure, 2 = high failure
    
    return df    
```
